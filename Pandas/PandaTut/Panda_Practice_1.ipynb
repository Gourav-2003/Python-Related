```python
# Applying Functions with .apply()
def RadiusCategory(row):
    if row['radius_se'] < 0.405172 :
        return "Below Avg"
    else :
        return "Above Avg"
    
df['radius_se_category'] = df.apply(RadiusCategory , axis = 1)
df[['radius_se' , 'radius_se_category']]

```
